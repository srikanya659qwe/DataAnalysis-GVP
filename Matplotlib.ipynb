#### pyplot():
    - pyplot is a sub module in matplotlib
    - it contains collection of functions to provide high level interface to design plots in matplotlib

```python
import matplotlib.pyplot as plt
```

#### how to read an image in pyplot:

```python
print(dir(plt))
```
