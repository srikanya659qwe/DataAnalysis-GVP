#### lineplot or linear plot:
            - plt.plot()

```python
x_axis=[56,78,90,34]
y_axis=[23,45,70,95]
plt.plot(x_axis,y_axis,label="Graph")
plt.title("Linear Graph")
plt.xlabel("X_Axis")
plt.ylabel("Y_Axis")
plt.legend(loc="upper right")
plt.show()
```

```python
import numpy as np
```

```python
a=np.arange(5)
print(a)
plt.plot(a,a+0.5,color='green')
ab=plt.axes()
print(ab)
ab.set_facecolor("red")
plt.plot(a,a+1.0,color='',linestyle='-')
plt.plot(a,a+1.5,color='blue',linestyle='--')
plt.plot(a,a+2.0,color='maroon',linestyle=':')
plt.plot(a,a+2.5,color='yellow',linestyle='-.')
plt.show()
```

```python
l=np.linspace(1,10,30)
plt.plot(l,np.sin(l),marker='o',markersize=8,markerfacecolor='g')
plt.plot(l,np.cos(l),marker='D',markersize=8,markerfacecolor='r')
plt.show()
```

#### Barplot:
        - plt.bar()

```python
year=['2024 Jan','2023 Jan','2022 Jan','2021 Jan']
gold_price=[87000,85000,80000,78000]
silver_price=[10000,15000,12000,8000]
plt.bar(year,gold_price,color='g',width=0.34)
plt.bar(year,silver_price,color='yellow',width=0.23)
plt.show()
```

```python
y=np.arange(5)
plt.bar(y,[23,56,89,67,90],width=0.34,color='m')
plt.bar(y+0.5,[12,50,46,87,60],width=0.3,color='r')
plt.show()
```

#### horizotal bar:
        plt.barh()

```python
plt.barh(y+0.5,[12,50,46,87,60],color='r')
plt.barh(y,[23,56,89,67,90],color='m')
```

#### scatter plot()
        - plt.scatter()

```python
plt.scatter([5,9],[4,7])
```

```python
r=np.random.randint(500,1500,1000)
r1=np.random.randint(1500,2500,1000)
plt.scatter(r,r1,color='g')
plt.show()
```

```python
r=np.random.randint(500,1500,1000)
r1=np.random.randint(1500,2500,1000)
plt.scatter(r,r1,color='g',marker='+')
plt.show()
```

#### boxplot():
        plt.boxplot()

```python
r2=np.random.randint(100,1000,500)
plt.boxplot(r2,notch=True)
plt.show()
```

#### piechart()
        - plt.pie()

```python
games=['cricket','koko','football','ludo','chess']
score=[95,87,65,54,43]
plt.pie(score,labels=games,startangle=90,labeldistance=1.5,shadow=True)
plt.show()
```
